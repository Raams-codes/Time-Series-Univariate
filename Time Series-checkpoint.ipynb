```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\DELL\\Desktop\\DS\\ML\\Time series\\AirPassengers.csv")
```

```python
df.head()
```

```python
df["#Passengers_diff"]=df["#Passengers"]-df["#Passengers"].shift(1)
```

```python
df["asd"]=df["#Passengers"].rolling(window=30).mean()
```

```python
df["asf"]=df["#Passengers"].rolling(window=30).std()
```

```python
data=df["#Passengers"].plot()
diff=df["#Passengers_diff"].dropna().plot()
mean=plt.plot(df["asd"],label="mean")
std=plt.plot(df["asf"],label="std")
plt.legend(loc="best")
plt.show()
```

```python
df["adj_log"]=np.log(df["#Passengers"])
```

```python
df["adj_sqrt"]=np.sqrt(df["#Passengers"])
```

```python
df["adj_cbrt"]=np.cbrt(df["#Passengers"])
```

```python
df["adj_log"].plot()
df["adj_sqrt"].plot()
df["adj_cbrt"].plot()
plt.legend()
plt.show()
```
